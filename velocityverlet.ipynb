```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
h = 0.1                        # time step
n=10/h                         # number of steps
n=int(n)
x = np.zeros(n+1)              # array of positions
p = np.zeros(n+1)              # array of momenta
f = np.zeros(n+1)              # array of forces
t = np.linspace(0,10,n)
p[0] = 1e-3                    # initial momenta
m = 1                          # mass of the particle
k = 0.1                        # boltzman constant
```

```python
"""calculates the new positions and momenta using half time step, involves a force calculation"""


def verlet_method(n,h):
    
    for i in range(n):
        x[i+1] = x[i] + p[i]*h/m + 0.5*f[i]/m*h**2  # update position
        
        f[i+1] = -k*x[i+1]                          # update force
        
        p[i+1] = p[i] + 0.5*h* (f[i]+f[i+1])        # update momentum
        
        
```

```python
verlet_method(n,h)
```

```python
plt.title("displacement x(t), verlet method")
plt.plot(x,label = "h = 10^-3")
plt.legend(loc="upper left")
```

```python
plt.title("momentum p(t), verlet method")
plt.plot(p,label = "h = 10^-1")
plt.legend(loc="upper left")
```

```python
E = np.zeros(n)
def verlet_energy(n,h):
    for i in range(n):
        E[i] = 0.5*p[i]**2/m + 0.5*k*x[i]**2      # total energy at each timestep
```

```python
verlet_energy(n,h)
```

```python
plt.title("phase space trajectory verlet method")
plt.plot(x,p)
plt.xlabel("Displacement x(t)")
plt.ylabel("Momentum p(t)")
```

```python
plt.title("Energy, verlet method")
plt.plot(t, E)
plt.ylim(0, 10e-7)
plt.show()
```

```python
e = E - E[0]
```

```python
plt.title("energy deviation, verlet method")

plt.plot(t,e, label = "h= 10^-3")
plt.legend(loc="upper left")
plt.show()
```

```python
plt.title("energy deviation, verlet method")
plt.plot(e, label = "h= 10^-1")
plt.legend(loc="upper left")

plt.show()
```

```python
e1 = e/h
plt.title("energy deviation/ timestep")
plt.plot(t, e1)
plt.show()
```

```python
e2 = e/(h**2)
plt.title("energy deviation/timestep^2")
plt.plot(t, e2)
plt.show()
```

```python
e1 = e/h
e2 = e/h**2
plt.figure(figsize=(12,2))
figure, axis = plt.subplots(1,2)
figure.suptitle("verlet method, h= 10^-1")
axis[0].plot(t,e1)
axis[0].set_title("energy deviation/ timestep")
axis[1].plot(t,e2)
axis[1].set_title("energy deviation/ timestep^2")
```
